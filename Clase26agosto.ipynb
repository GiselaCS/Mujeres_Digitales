```python
# Standard
from bokeh.io import output_notebook, show, reset_output

# Other
import bokeh
from bokeh.plotting import figure

import numpy as np
import pandas as pd
```

```python
#from sklearn import datasets
#iris = datasets.load_iris()

from bokeh.sampledata import iris
iris = iris.flowers
iris.head()
```

```python
output_notebook() # To show in notebook cells
```

```python
line_plot = figure(plot_width=500, plot_height=300, title='Line plot', 
                   x_axis_label='x', y_axis_label='y')
line_plot.line(iris.index, iris['sepal_length'], legend='Line', line_width=2)
show(line_plot)
```

```python
from bokeh.models import HoverTool

line_plot = figure(plot_width=500, plot_height=300, title='Line plot', 
                   x_axis_label='x', y_axis_label='y')
line_plot.line(iris.index, iris['sepal_length'], legend='Line', line_width=2)
line_plot.add_tools(HoverTool())

show(line_plot)
```
