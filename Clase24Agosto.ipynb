```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pk = pd.read_csv('pokemon_data.csv')
```

```python
data= pk.iloc[:, 9]
bins= 10 #numero de particiones en los que se quiere hacer 
plt.hist(data,bins, histtype='bar', rwidth=0.8, color="pink")
plt.title('Histogram Speed')
```

```python
dt = pk.iloc[:,1]
g= pk.iloc[:,10]
bins = 2
plt.hist(dt,g,bins, histtype='bar', rwidth=0.5, color="green")
plt.title('Histogram')
plt.xlabel("Names"), plt.xticks(rotation=60)
plt.ylabel("Generation")
```

```python
dt = pk.iloc[:,1]
g= pk.iloc[:,10]
bins = 2
plt.hist(dt,g,bins, histtype='bar', rwidth=0.3, color="gray")
plt.title('Histogram')
plt.xlabel("Nombres"), plt.xticks(rotation=60)
plt.ylabel("Generacion")
```
